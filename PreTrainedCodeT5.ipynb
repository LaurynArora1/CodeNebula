```python
from transformers import RobertaTokenizer, T5ForConditionalGeneration
import nltk
from nltk.translate.bleu_score import sentence_bleu
import pandas as pd
tokenizer = RobertaTokenizer.from_pretrained('Salesforce/codet5-base')
model = T5ForConditionalGeneration.from_pretrained('Salesforce/codet5-base-multi-sum')
data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/preprocessed_data (1).csv")
df=pd.DataFrame(columns=["code","existing_summary","generated_summary","bleu"])
b=[]
for i in range(len(data)):
  text = data.iloc[i,3]
  input_ids = tokenizer(text, return_tensors="pt").input_ids
  generated_ids = model.generate(input_ids, max_length=20)
  reference=tokenizer.decode(generated_ids[0], skip_special_tokens=True)
  # this prints: "Convert a SVG string to a QImage."
  candidate = data.iloc[i,5]
  # Tokenize the sentences
  reference_tokens = [sentence.split() for sentence in reference]
  candidate_tokens = candidate.split()
  # Calculate BLEU score
  bleu_score = sentence_bleu(reference_tokens, candidate_tokens)
  # Print the BLEU score
  b.append(bleu_score)
  new_row={"code":text, "existing_summary":candidate, "generated_summary":reference, "bleu":bleu_score}
  df = df.append(new_row, ignore_index=True)
print("Sum with zeros",sum(b)/len(b))
s=0
c=0
for i in b:
  if i!=0:
    s+=i
    c+=1
print("Sum without zeros",s/c)
df.to_csv("output_codet5_2.csv")
```
